```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# <center> precio_energia

```python
df1 = pd.read_excel('precio_energia.xlsx')
print(df1.head())
```

```python
print('Primeras filas')
print(df1.head())

print('')
print('Descripción')
print(df1.describe())

print('')
print('Información de los datos')
print(df1.info())
```

```python
print(df1.columns)
```

```python
# Análisis de series temporales

# Convertir la columna en un índice de tiempo
df1.rename(columns={' Date ': 'Date'}, inplace=True)  # Ajusta el nombre según sea necesario
df1.set_index('Date', inplace=True)

# Plot de una serie temporal
df1['precio_nutresa'].plot()
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.show()
```

# <center> materia_prima

```python
df2 = pd.read_excel('materia_prima.xlsx')
print(df2.head())
```

```python
print(df2.columns)
```

```python
print('')
print('Descripción')
print(df2.describe())

print('')
print('Información de los datos')
print(df2.info())
```

```python
# Análisis de series temporales

# Plot de una serie temporal
df2['cacao_nal'].plot()
plt.xlabel('Fecha')
plt.ylabel('cacao_nal')
plt.show()

# Plot de una serie temporal
df2['Precio_cacao_und'].plot()
plt.xlabel('Fecha')
plt.ylabel('Precio del cacao por unidad')
plt.show()

# Plot de una serie temporal
df2['cantidad_cacao'].plot()
plt.xlabel('Fecha')
plt.ylabel('Cantidad de cacao')
plt.show()
```

```python
# Seleccionar las columnas 'Date', 'cacao_nal', 'precio_cacao_und' y 'cantidad_cacao'
cantidades = df2[['Date', 'cacao_nal', 'cafe_nal', 'cerdo_nal', 'res_nal']]

# Mostrar las primeras filas del DataFrame seleccionado para verificar
print(cantidades.head())

# Plot de las series temporales
plt.figure(figsize=(10, 6))

# Graficar cada serie temporal
plt.plot(cantidades.index, cantidades['cacao_nal'], label='Cantidad de Cacao Nacional')
plt.plot(cantidades.index, cantidades['cafe_nal'], label='Cantidad de Café Nacional')
plt.plot(cantidades.index, cantidades['cerdo_nal'], label='Cantidad de Cerdo Nacional')
plt.plot(cantidades.index, cantidades['res_nal'], label='Cantidad de Res Nacional')

# Añadir título y etiquetas
plt.title('Materia prima') # No sé esta columna de que es
plt.xlabel('Fecha')
plt.ylabel('Valores')
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
# Seleccionar las columnas 'Date', 'cacao_nal', 'precio_cacao_und' y 'cantidad_cacao'
df_selected = df2[['Date', 'Precio_cacao_und', 'Precio_cafe', 'Precio_carne_cerdo', 'Precio_carne_res']]

# Mostrar las primeras filas del DataFrame seleccionado para verificar
print(df_selected.head())

# Plot de las series temporales
plt.figure(figsize=(10, 6))

# Graficar cada serie temporal
plt.plot(df_selected.index, df_selected['Precio_cacao_und'], label='Precio del Cacao Nacional')
plt.plot(df_selected.index, df_selected['Precio_cafe'], label='Precio del Café Nacional')
plt.plot(df_selected.index, df_selected['Precio_carne_cerdo'], label='Precio de Cerdo Nacional')
plt.plot(df_selected.index, df_selected['Precio_carne_res'], label='Precio de Res Nacional')

# Añadir título y etiquetas
plt.title('Cantidades de materia prima')
plt.xlabel('Fecha')
plt.ylabel('Valores')
plt.legend()

# Mostrar la gráfica
plt.show()
```
