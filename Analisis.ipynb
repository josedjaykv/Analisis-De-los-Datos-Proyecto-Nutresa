```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# <center> precio_energia

```python
df1 = pd.read_excel('precio_energia.xlsx')
print(df1.head())
```

```python
print('Primeras filas')
print(df1.head())

print('')
print('Descripción')
print(df1.describe())

print('')
print('Información de los datos')
print(df1.info())
```

```python
print(df1.columns)
```

```python
# Análisis de series temporales

# Convertir la columna en un índice de tiempo
df1.rename(columns={' Date ': 'Date'}, inplace=True)  # Ajusta el nombre según sea necesario
df1.set_index('Date', inplace=True)

# Plot de una serie temporal
df1['precio_nutresa'].plot()
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.show()
```
