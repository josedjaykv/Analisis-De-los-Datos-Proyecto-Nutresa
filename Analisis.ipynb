```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# <center> precio_energia

# <center> materia_prima

```python
df2 = pd.read_excel('Data/materia_prima.xlsx')
print(df2.head())
```

```python
print(df2.columns)
```

```python
print('')
print('Descripción')
print(df2.describe())

print('')
print('Información de los datos')
print(df2.info())
```

```python
# Análisis de series temporales

# Plot de una serie temporal
df2['cacao_nal'].plot()
plt.xlabel('Fecha')
plt.ylabel('cacao_nal')
plt.show()

# Plot de una serie temporal
df2['Precio_cacao_und'].plot()
plt.xlabel('Fecha')
plt.ylabel('Precio del cacao por unidad')
plt.show()

# Plot de una serie temporal
df2['cantidad_cacao'].plot()
plt.xlabel('Fecha')
plt.ylabel('Cantidad de cacao')
plt.show()
```

```python
# Seleccionar las columnas 'Date', 'cacao_nal', 'precio_cacao_und' y 'cantidad_cacao'
cantidades = df2[['Date', 'cacao_nal', 'cafe_nal', 'cerdo_nal', 'res_nal']]

# Mostrar las primeras filas del DataFrame seleccionado para verificar
print(cantidades.head())

# Plot de las series temporales
plt.figure(figsize=(10, 6))

# Graficar cada serie temporal
plt.plot(cantidades.index, cantidades['cacao_nal'], label='Cantidad de Cacao Nacional')
plt.plot(cantidades.index, cantidades['cafe_nal'], label='Cantidad de Café Nacional')
plt.plot(cantidades.index, cantidades['cerdo_nal'], label='Cantidad de Cerdo Nacional')
plt.plot(cantidades.index, cantidades['res_nal'], label='Cantidad de Res Nacional')

# Añadir título y etiquetas
plt.title('Materia prima') # No sé esta columna de que es
plt.xlabel('Fecha')
plt.ylabel('Valores')
plt.legend()

# Formatear los valores del eje y para que se muestren exactamente
plt.ticklabel_format(axis='y', style='plain')

# Mostrar la gráfica
plt.show()
```

# <center> Energia diaria


```python
df_energia_diaria= pd.read_excel('data/energia_diaria.xlsx')
print(df_energia_diaria.head())
```

```python
print('Primeras filas')
print(df_energia_diaria.head())

print('')
print('Descripción')
print(df_energia_diaria.describe())

print('')
print('Información de los datos')
print(df_energia_diaria.info())
```

```python
print("Columnas originales:", df_energia_diaria.columns)
```

```python
# Plot de una serie temporal
df_energia_diaria['Total_Nutresa'].plot()
plt.xlabel('Fecha')
plt.ylabel('Total')
plt.show()
```

```python
# Seleccionar los primeros 300 datos
df_energia_100 = df_energia_diaria.head(366)

# Plot de los primeros 100 datos de la serie temporal
plt.figure(figsize=(10, 6))
df_energia_100['Total_Nutresa'].plot()
plt.xlabel('Fecha')
plt.ylabel('Total Nutresa')
plt.title('Serie Temporal de los Primeros 366 Datos de Total Nutresa')
plt.grid(True)
plt.show()
```

```python
# Seleccionar los primeros 300 datos
df_energia_segmento = df_energia_diaria.iloc[366:733]

# Plot de los primeros 100 datos de la serie temporal
plt.figure(figsize=(10, 6))
df_energia_segmento['Total_Nutresa'].plot()
plt.xlabel('Fecha')
plt.ylabel('Total Nutresa')
plt.title('Serie Temporal del 2016 Datos de Total Nutresa')
plt.grid(True)
plt.show()
```
